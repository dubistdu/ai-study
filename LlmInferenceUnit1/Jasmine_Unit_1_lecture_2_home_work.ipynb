```python
# Import the Hugging Face Transformers library's classes:
# - AutoTokenizer: Automatically loads the correct tokenizer for the given pre-trained model.
#   The tokenizer converts text (strings) into numerical token IDs that the model can understand,
#   and vice versa (decoding token IDs back to text).
#
# - AutoModelForCausalLM: Automatically loads a pre-trained language model suitable for causal
#   language modeling (i.e., generating text). "Causal LM" models predict the next token in a sequence,
#   making them ideal for tasks like text generation, chatbots, or autocompletion.
#
# These classes simplify the process of leveraging state-of-the-art NLP models by handling the
# underlying model architecture and tokenization specifics based on a model identifier.
from transformers import AutoTokenizer, AutoModelForCausalLM


# -------------------------------------------------------------------------------------
# LOAD THE TOKENIZER AND MODEL
# -------------------------------------------------------------------------------------
# 1. "AutoTokenizer.from_pretrained(...)":
#    - AutoTokenizer is a convenient wrapper that automatically selects the
#      appropriate tokenizer class for a given model name.
#    - load "facebook/opt-125m" is a smaller version of Meta's OPT model series.
#      The "125m" indicates ~125 million parameters.
#    - The tokenizer is responsible for converting raw text strings (prompts)
#      into tokens (integers) that the model can process.

tokenizer = AutoTokenizer.from_pretrained("facebook/opt-125m")

# 2. Load the Model "AutoModelForCausalLM.from_pretrained(...)":
#    - AutoModelForCausalLM is a wrapper that automatically sets up the
#      specified model for "causal language modeling" (i.e., generating text).
#    - load "facebook/opt-125m" corresponds to the same model family as the tokenizer.
#    - This call downloads (if needed) and initializes the model weights.

model = AutoModelForCausalLM.from_pretrained("facebook/opt-125m")

# -------------------------------------------------------------------------------------
# Create A RESPONSE
# -------------------------------------------------------------------------------------
# We now create a prompt (input string) for the model to generate a continuation.
# The prompt "What is the capital of France?" is a simple test question.
# You can change this prompt to explore different kinds of responses or tasks.
prompt = "What is love"  # Modify prompt here

# -------------------------------------------------------------------------------------
# TOKENIZATION
# -------------------------------------------------------------------------------------
# "tokenizer(prompt, return_tensors='pt')" encodes the prompt into the proper format:
#  - 'pt' indicates we're returning PyTorch tensors.
#  - The tokenizer will:
#       * Convert each token (sub-word piece) to an integer ID.
#       * Build an input dictionary with attention masks and possibly other metadata.
inputs = tokenizer(prompt, return_tensors="pt")

# -------------------------------------------------------------------------------------
# Mess with Max_lenth GENERATION
# -------------------------------------------------------------------------------------
# "model.generate" uses a causal language modeling head to predict the next tokens:
#  - "max_length=50" sets an upper bound on the generated sequence length.
#  - We pass "**inputs" (the tokenized prompt) as keyword arguments to generate.
#  - The model returns a sequence of token IDs.
outputs = model.generate(**inputs, max_length=50)

# -------------------------------------------------------------------------------------
# DECODING THE OUTPUT
# -------------------------------------------------------------------------------------
# "tokenizer.decode" converts the sequence of token IDs back into a human-readable string.
#  - "outputs[0]" grabs the first (and in this case, only) generated sequence from the model.
#  - "skip_special_tokens=True" removes special tokens like <EOS> or <BOS>.
# The printed result is the model’s response to our prompt.
print(tokenizer.decode(outputs[0], skip_special_tokens=True))
```

# Exercise 3: Tweaking Generation Parameters. In this Exercise, you need to provide value in for loop and debug the code in such a way that code uses for loop values in the code after each for loop
Objective:

    Experiment with different generation parameters:
        max_length or max_token or max_new_token: Determines the maximum number of tokens in the generated sequence.
        temperature: Controls the randomness of predictions (lower values → more deterministic output, higher values → more randomness).
        top_k: Restricts the model to sample from only the top K most likely tokens.
        top_p (nucleus sampling): Samples from the smallest possible set of tokens whose cumulative probability exceeds a threshold p.
    For a fixed prompt, generate text using different parameter values.
    Compare and document how changes in these parameters affect the length, randomness, and overall style of the generated output.

```python
from transformers import AutoTokenizer, AutoModelForCausalLM

# Load the tokenizer and model (using a smaller model here for quick experimentation)
tokenizer = AutoTokenizer.from_pretrained("facebook/opt-125m")
model = AutoModelForCausalLM.from_pretrained("facebook/opt-125m")

# -------------------------------------------------------------------------------
# STEP 1: DEFINE A FIXED PROMPT
# -------------------------------------------------------------------------------
# A fixed prompt ensures that any changes in the output are due to parameter adjustments.
prompt = "Hello Darkness, my old friend. I've come to talk with you again. Because the vision"   # SET A PROMPT Try writing any incomplete sentence. For example: Once upon a Time, I love, ....
inputs = tokenizer(prompt, return_tensors="pt")

# -------------------------------------------------------------------------------
# STEP 2: EXPERIMENT WITH max_length
# -------------------------------------------------------------------------------
# The max_length parameter controls the maximum number of tokens in the generated sequence.
# Trying different values will show you how much the output grows or truncates.
print("=== Experimenting with max_length ===")
for length in [30,50,70]:
    outputs = model.generate(**inputs, max_length=50)
    generated_text = tokenizer.decode(outputs[0], skip_special_tokens=True)
    print(f"\nMax Length = {length}:")
    print(generated_text)

# -------------------------------------------------------------------------------
# STEP 3: EXPERIMENT WITH temperature
# -------------------------------------------------------------------------------
# Temperature influences randomness in the model's predictions:
# - Lower temperature (e.g., 0.7) makes the output more deterministic and focused.
# - Higher temperature (e.g., 1.3) increases randomness and creativity.
print("\n=== Experimenting with temperature ===")
for temp in []:
    outputs = model.generate(**inputs, max_length=50, temperature=0.7)
    generated_text = tokenizer.decode(outputs[0], skip_special_tokens=True)
    print(f"\nTemperature = {temp}:")
    print(generated_text)

# -------------------------------------------------------------------------------
# STEP 4: EXPERIMENT WITH top_k
# -------------------------------------------------------------------------------
# The top_k parameter limits the model's sampling to the top K tokens at each step.
# Lower values (e.g., top_k=10) can lead to more focused (but sometimes repetitive) text,
# while higher values allow more diversity in the generated text.
print("\n=== Experimenting with top_k ===")
for top_k in []:
    outputs = model.generate(**inputs, max_length=30, top_k=50)
    generated_text = tokenizer.decode(outputs[0], skip_special_tokens=True)
    print(f"\nTop K = {top_k}:")
    print(generated_text)

# -------------------------------------------------------------------------------
# STEP 5: (OPTIONAL) EXPERIMENT WITH top_p (NUCLEUS SAMPLING)
# -------------------------------------------------------------------------------
# The top_p parameter (nucleus sampling) allows the model to consider the smallest
# set of tokens whose cumulative probability exceeds p.
# This provides another way to control diversity in the output.
print("\n=== Experimenting with top_p ===")
for top_p in []:
    outputs = model.generate(**inputs, max_length=50, top_p=50)
    generated_text = tokenizer.decode(outputs[0], skip_special_tokens=True)
    print(f"\nTop P = {top_p}:")
    print(generated_text)
```

```python
from transformers import AutoTokenizer, AutoModelForCausalLM

# -------------------------------------------------------------------------------------
# LOAD THE MODEL AND TOKENIZER
# -------------------------------------------------------------------------------------
# We are using a pre-trained model ("facebook/opt-125m") from Hugging Face.
# AutoTokenizer converts input text into tokens that the model can process.
# AutoModelForCausalLM is used for causal language modeling (text generation).
# The model "facebook/opt-125m" is relatively small, making it suitable for quick experiments.
tokenizer = AutoTokenizer.from_pretrained("facebook/opt-125m")
model = AutoModelForCausalLM.from_pretrained("facebook/opt-125m", device_map="auto")

# -------------------------------------------------------------------------------------
# DEFINE TWO DIFFERENT PROMPTS
# -------------------------------------------------------------------------------------
# The goal of this exercise is to demonstrate the importance of prompt engineering.
# We will compare the outputs of a simple prompt versus an engineered prompt.
#
# Simple prompt:
#   A straightforward, minimal input asking for information.
simple_prompt = ""  # Ask a straight forward question. For example: What are stocks and recommend me some.

# Engineered prompt:
#   A more detailed and context-rich prompt designed to encourage the model
#   to provide a more comprehensive and nuanced response.
engineered_prompt = (
    ""
    ""
    ""
) #Related that straight forward question with 3 of the prompt engineering template from notion. For example: 1) Act as if you are a HedgeFund Advisor, 2) Act as if you are a expert in Quant Finanace, 3) I am going to provide a template for your output. Use the format: 'Dear [CUSTOMER_NAME], thank you for your asking advice of [INPUT]. I would recommend these 3 [STOCKS_NAME]'.

# -------------------------------------------------------------------------------------
# TOKENIZE THE PROMPTS
# -------------------------------------------------------------------------------------
# The tokenizer converts the text inputs into token IDs.
# The 'return_tensors="pt"' parameter converts the tokens into PyTorch tensors.
simple_inputs = tokenizer(simple_prompt, return_tensors="pt")
engineered_inputs = tokenizer(engineered_prompt, return_tensors="pt")

# -------------------------------------------------------------------------------------
# GENERATE TEXT FOR BOTH PROMPTS
# -------------------------------------------------------------------------------------
# We use model.generate() to generate a continuation for each prompt.
# The max_length parameter limits the length of the generated text.
simple_outputs = model.generate(**simple_inputs, max_length=50)
engineered_outputs = model.generate(**engineered_inputs, max_length=50)

# -------------------------------------------------------------------------------------
# DECODE THE GENERATED OUTPUTS
# -------------------------------------------------------------------------------------
# Convert the token IDs back into human-readable text.
# The option 'skip_special_tokens=True' removes any special tokens added during generation.
simple_text = tokenizer.decode(simple_outputs[0], skip_special_tokens=True)
engineered_text = tokenizer.decode(engineered_outputs[0], skip_special_tokens=True)

# -------------------------------------------------------------------------------------
# PRINT AND COMPARE THE OUTPUTS
# -------------------------------------------------------------------------------------
print("=== Simple Prompt Output ===")
print(simple_text)
print("\n=== Engineered Prompt Output ===")
print(engineered_text)

# -------------------------------------------------------------------------------------
# CONTEXT & ANALYSIS
# -------------------------------------------------------------------------------------
# In this exercise, you should observe how the style and content of the generated text
# differ based on the prompt:
#
# 1. Simple Prompt:
#    - The output is likely to be short and direct, simply naming "Paris" or providing minimal context.
#
# 2. Engineered Prompt:
#    - The output may be more detailed and comprehensive, potentially discussing historical,
#      cultural, and political factors that establish Paris as the capital of France.
#
# By comparing these outputs, you'll understand the importance of prompt engineering:
# - A well-crafted prompt can guide the model to produce richer and more informative responses.
# - This is crucial for applications where the quality and depth of information are important,
#   such as in chatbots, content generation, or educational tools.
```
